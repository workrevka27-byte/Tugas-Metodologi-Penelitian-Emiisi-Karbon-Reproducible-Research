```r
# Carbon Emission Reproducible Research
## Notebook 02 - Perhitungan Vehicle Kilometer Travelled (VKT) dan Emisi Karbon

Notebook ini bertujuan untuk:

- Mengimpor data hasil pembersihan
- Menghitung Vehicle Kilometer Travelled (VKT)
- Menghitung emisi karbon kendaraan
- Melakukan validasi hasil perhitungan
- Menghasilkan dataset yang siap dianalisis pada Notebook 03

**Input**
- output/data_clean.csv

**Output**
- output/hasil_emisi.csv
```

```r
# ============================================================
# Project : Carbon Emission Reproducible Research
# File    : 02_perhitungan_emisi.R
# Author  : Revka Saputro
# Date    : 09 Juli 2026
#
# Purpose :
# Menghitung Vehicle Kilometer Travelled (VKT),
# menghitung emisi karbon kendaraan,
# melakukan validasi hasil,
# dan menghasilkan dataset hasil perhitungan.
# ============================================================
```

```r
# ============================================================
# 1. LOAD PACKAGE
# ============================================================

library(readr)
library(dplyr)

set.seed(123)
```

```r
data <- read.csv(
  "output/data_clean.csv",
  stringsAsFactors = FALSE
)
```

```r
# ============================================================
# 2. IMPORT DATA
# ============================================================

data <- read_csv(
  "output/data_clean.csv",
  show_col_types = FALSE
)

cat("Data berhasil diimpor.\n")
```

```r
# ============================================================
# 3. CEK DATA
# ============================================================

head(data)

str(data)

summary(data)
```

```r
# ============================================================
# 4. PERHITUNGAN VEHICLE KILOMETER TRAVELLED
# ============================================================

data <- data %>%
  mutate(
    vkt = volume * panjang_jalan
  )

head(data)
```

```r
# ============================================================
# 5. PERHITUNGAN EMISI KARBON
# ============================================================

data <- data %>%
  mutate(
    emisi_g = vkt * faktor_emisi,
    emisi_kg = emisi_g / 1000,
    emisi_ton = emisi_kg / 1000
  )

head(data)
```

```r
# ============================================================
# 6. VALIDASI HASIL PERHITUNGAN
# ============================================================

summary(data$vkt)

summary(data$emisi_g)

summary(data$emisi_kg)

summary(data$emisi_ton)
```

```r
# ============================================================
# 7. VALIDASI NILAI NEGATIF
# ============================================================

sum(data$vkt < 0)

sum(data$emisi_g < 0)

sum(data$emisi_kg < 0)

sum(data$emisi_ton < 0)
```

```r
# ============================================================
# 8. STATISTIK DESKRIPTIF
# ============================================================

cat("Jumlah observasi :", nrow(data), "\n")

cat("Total VKT :", sum(data$vkt), "\n")

cat("Total Emisi (gram) :", sum(data$emisi_g), "\n")

cat("Total Emisi (kg) :", sum(data$emisi_kg), "\n")

cat("Total Emisi (ton) :", sum(data$emisi_ton), "\n")
```

```r
# ============================================================
# 9. EMISI BERDASARKAN JENIS KENDARAAN
# ============================================================

emisi_kendaraan <- data %>%
  group_by(jenis_kendaraan) %>%
  summarise(
    Total_VKT = sum(vkt),
    Total_Emisi = sum(emisi_ton)
  )

print(emisi_kendaraan)
```

```r
# ============================================================
# 10. EMISI BERDASARKAN RUAS JALAN
# ============================================================

emisi_ruas <- data %>%
  group_by(ruas_jalan) %>%
  summarise(
    Total_VKT = sum(vkt),
    Total_Emisi = sum(emisi_ton)
  ) %>%
  arrange(desc(Total_Emisi))

print(emisi_ruas)
```

```r
# ============================================================
# 11. KONTRIBUSI EMISI
# ============================================================

emisi_ruas <- emisi_ruas %>%
  mutate(
    Persentase = 100 * Total_Emisi / sum(Total_Emisi)
  )

print(emisi_ruas)
```

```r
# ============================================================
# 12. EXPORT DATA
# ============================================================

write_csv(
  data,
  "output/hasil_emisi.csv"
)

write_csv(
  emisi_ruas,
  "output/ringkasan_ruas.csv"
)

write_csv(
  emisi_kendaraan,
  "output/ringkasan_kendaraan.csv"
)

cat("Seluruh hasil berhasil disimpan.\n")
```
