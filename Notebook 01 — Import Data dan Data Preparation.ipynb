```r
# ============================================================
# 1. LOAD PACKAGE
# ============================================================

library(readr)

set.seed(123)
```

```r
# ============================================================
# 2. MEMBUAT FOLDER OUTPUT
# ============================================================

if(!dir.exists("output")){
  dir.create("output")
}

cat("Folder output siap digunakan.\n")
```

```r
# ============================================================
# 3. IMPORT DATA
# ============================================================

data <- read_csv(
  "data/data_emisi.csv",
  show_col_types = FALSE
)

cat("Data berhasil diimpor.\n")
```

```r
data <- read.csv(
  "data/data_emisi.csv",
  header = TRUE,
  stringsAsFactors = FALSE
)
```

```r
# ============================================================
# 4. MENAMPILKAN DATA
# ============================================================

head(data)

tail(data)
```

```r
# ============================================================
# 5. STRUKTUR DATA
# ============================================================

str(data)

summary(data)

names(data)

dim(data)
```

```r
# ============================================================
# 6. TIPE DATA
# ============================================================

sapply(data, class)
```

```r
# ============================================================
# 7. CEK MISSING VALUE
# ============================================================

colSums(is.na(data))
```

```r
# ============================================================
# 9. RINGKASAN VARIABEL
# ============================================================

summary(data$volume)

summary(data$panjang_jalan)

summary(data$faktor_emisi)
```

```r
# ============================================================
# 10. VALIDASI NILAI
# ============================================================

sum(data$volume < 0)

sum(data$panjang_jalan < 0)

sum(data$faktor_emisi < 0)
```

```r
# ============================================================
# 15. EXPORT DATA CLEAN
# ============================================================

write_csv(
  data,
  "output/data_clean.csv"
)

cat("Data bersih berhasil disimpan.\n")
```
