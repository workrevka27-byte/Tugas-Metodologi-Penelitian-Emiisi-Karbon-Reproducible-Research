```r
# Carbon Emission Reproducible Research
## Notebook 03 - Analisis Statistik dan Visualisasi Emisi Karbon

Notebook ini bertujuan untuk menganalisis hasil perhitungan emisi karbon kendaraan bermotor menggunakan statistik deskriptif dan visualisasi data.

### Input
- output/hasil_emisi.csv

### Output
- output/ringkasan_ruas.csv
- output/ringkasan_kendaraan.csv
- output/grafik_emisi_ruas.png
- output/grafik_emisi_kendaraan.png
- output/grafik_kontribusi_ruas.png
```

```r
# ============================================================
# Project : Carbon Emission Reproducible Research
# File    : 03_analisis_visualisasi.R
# Author  : Revka Saputro
# Date    : 09 Juli 2026
#
# Purpose :
# Melakukan analisis statistik deskriptif,
# menganalisis emisi karbon berdasarkan ruas jalan
# dan jenis kendaraan,
# membuat visualisasi,
# serta menghasilkan ringkasan hasil penelitian.
# ============================================================
```

```r
# ============================================================
# 1. LOAD PACKAGE
# ============================================================

library(readr)
library(dplyr)
library(ggplot2)

set.seed(123)
```

```r
# ============================================================
# 2. IMPORT DATA
# ============================================================

data <- read_csv(
  "output/hasil_emisi.csv",
  show_col_types = FALSE
)

cat("Data berhasil diimpor.\n")
```

```r
# ============================================================
# 3. CEK DATA
# ============================================================

head(data)

str(data)

summary(data)
```

```r
# ============================================================
# 4. STATISTIK DESKRIPTIF
# ============================================================

summary(data$vkt)

summary(data$emisi_ton)

cat("Jumlah Observasi :", nrow(data), "\n")

cat("Jumlah Ruas :", length(unique(data$ruas_jalan)), "\n")

cat("Jenis Kendaraan :", length(unique(data$jenis_kendaraan)), "\n")
```

```r
# ============================================================
# 5. EMISI BERDASARKAN RUAS JALAN
# ============================================================

emisi_ruas <- data %>%
  group_by(ruas_jalan) %>%
  summarise(
    Volume = sum(volume),
    Total_VKT = sum(vkt),
    Total_Emisi = sum(emisi_ton)
  ) %>%
  arrange(desc(Total_Emisi))

print(emisi_ruas)
```

```r
# ============================================================
# 6. EMISI BERDASARKAN JENIS KENDARAAN
# ============================================================

emisi_kendaraan <- data %>%
  group_by(jenis_kendaraan) %>%
  summarise(
    Volume = sum(volume),
    Total_VKT = sum(vkt),
    Total_Emisi = sum(emisi_ton)
  )

print(emisi_kendaraan)
```

```r
# ============================================================
# 7. PERSENTASE KONTRIBUSI
# ============================================================

emisi_ruas <- emisi_ruas %>%
  mutate(
    Persentase = Total_Emisi/sum(Total_Emisi)*100
  )

print(emisi_ruas)
```

```r
# ============================================================
# 8. LIMA RUAS DENGAN EMISI TERBESAR
# ============================================================

head(emisi_ruas,5)
```

```r
# ============================================================
# 9. GRAFIK EMISI PER RUAS
# ============================================================

ggplot(
  emisi_ruas,
  aes(
    x = reorder(ruas_jalan, Total_Emisi),
    y = Total_Emisi
  )
)+

geom_col(fill="steelblue")+

coord_flip()+

labs(

title="Total Emisi Karbon Tiap Ruas Jalan",

x="Ruas Jalan",

y="Emisi (ton CO2)"

)+

theme_minimal()
```

```r
ggsave(

filename="output/grafik_emisi_ruas.png",

width=8,

height=5,

dpi=300
)
```

```r
# ============================================================
# 10. GRAFIK EMISI BERDASARKAN JENIS KENDARAAN
# ============================================================

ggplot(

emisi_kendaraan,

aes(

x=jenis_kendaraan,

y=Total_Emisi,

fill=jenis_kendaraan

)

)+

geom_col()+

labs(

title="Emisi Berdasarkan Jenis Kendaraan",

x="Jenis Kendaraan",

y="Emisi (ton CO2)"

)+

theme_minimal()
```

```r
ggsave(

filename="output/grafik_emisi_kendaraan.png",

width=6,

height=5,

dpi=300
)
```

```r
# ============================================================
# 11. KONTRIBUSI EMISI TIAP RUAS
# ============================================================

ggplot(

emisi_ruas,

aes(

x="",

y=Persentase,

fill=ruas_jalan

)

)+

geom_col(

width=1

)+

coord_polar("y")+

theme_void()+

labs(

title="Kontribusi Emisi Tiap Ruas"

)
```

```r
ggsave(

filename="output/grafik_kontribusi_ruas.png",

width=6,

height=6,

dpi=300
)
```

```r
# ============================================================
# 12. EXPORT RINGKASAN
# ============================================================

write_csv(

emisi_ruas,

"output/ringkasan_ruas.csv"

)

write_csv(

emisi_kendaraan,

"output/ringkasan_kendaraan.csv"

)
```
