```r
# ============================================================
# Project : Carbon Emission Reproducible Research
# File    : 04_export_output.R
# Author  : Revka Saputro
# Date    : 09 Juli 2026
#
# Purpose :
# Menggabungkan hasil analisis,
# mengekspor output penelitian,
# mendokumentasikan lingkungan komputasi,
# dan menghasilkan output akhir penelitian.
# ============================================================
```

```r
# ============================================================
# 1. LOAD PACKAGE
# ============================================================

library(readr)
library(dplyr)
library(writexl)

set.seed(123)
```

```r
# ============================================================
# 2. IMPORT HASIL ANALISIS
# ============================================================

hasil_emisi <- read_csv(
  "output/hasil_emisi.csv",
  show_col_types = FALSE
)

ringkasan_ruas <- read_csv(
  "output/ringkasan_ruas.csv",
  show_col_types = FALSE
)

ringkasan_kendaraan <- read_csv(
  "output/ringkasan_kendaraan.csv",
  show_col_types = FALSE
)

cat("Seluruh data berhasil diimpor.\n")
```

```r
# ============================================================
# 3. RINGKASAN PENELITIAN
# ============================================================

total_volume <- sum(hasil_emisi$volume)

total_vkt <- sum(hasil_emisi$vkt)

total_emisi <- sum(hasil_emisi$emisi_ton)

jumlah_ruas <- length(unique(hasil_emisi$ruas_jalan))

jumlah_jenis <- length(unique(hasil_emisi$jenis_kendaraan))
```

```r
# ============================================================
# 4. MENAMPILKAN RINGKASAN
# ============================================================

cat("\n==============================\n")

cat("RINGKASAN PENELITIAN\n")

cat("==============================\n")

cat("Jumlah Observasi :", nrow(hasil_emisi), "\n")

cat("Jumlah Ruas :", jumlah_ruas, "\n")

cat("Jumlah Jenis Kendaraan :", jumlah_jenis, "\n")

cat("Total Volume :", total_volume, "\n")

cat("Total VKT :", total_vkt, "\n")

cat("Total Emisi (ton CO2) :", total_emisi, "\n")
```

```r
# ============================================================
# 5. RUAS DENGAN EMISI TERBESAR
# ============================================================

head(ringkasan_ruas,5)
```

```r
# ============================================================
# 6. EMISI BERDASARKAN JENIS KENDARAAN
# ============================================================

ringkasan_kendaraan
```

```r
# ============================================================
# 7. EXPORT KE EXCEL
# ============================================================

write_xlsx(

list(

Hasil_Emisi = hasil_emisi,

Ringkasan_Ruas = ringkasan_ruas,

Ringkasan_Kendaraan = ringkasan_kendaraan

),

path = "output/ringkasan_penelitian.xlsx"

)

cat("File Excel berhasil dibuat.\n")
```

```r
# ============================================================
# 8. MENYIMPAN PACKAGE VERSION
# ============================================================

package_version <- data.frame(

Package = rownames(installed.packages()),

Version = installed.packages()[,"Version"]

)

write_csv(

package_version,

"output/package_version.csv"

)
```

```r
# ============================================================
# 9. SESSION INFO
# ============================================================

sink("output/sessionInfo_04.txt")

cat("Tanggal Analisis :", Sys.time(), "\n\n")

sessionInfo()

sink()
```
